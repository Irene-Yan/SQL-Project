# Refactoring

```python
def get_data(state):
    url = f"https://api.weather.gov/alerts/active?area={state}"
    r = requests.get(url)
    temp = pd.DataFrame([i["properties"] for i in r.json()["features"]])
    temp["State"] = state
    return(temp)
```

```python
df_all = pd.concat([get_data(state) for state in states])
```
